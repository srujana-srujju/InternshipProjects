```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.tiobe.com/tiobe-index/"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

language_elements = soup.select(".table.table-striped.table-top20 a")
rating_elements = soup.select(".table.table-striped.table-top20 .rating")

data = []

for language, rating in zip(language_elements, rating_elements):
    data.append([language.get_text(), rating.get_text()])

filename = "programming_languages.csv"

with open(filename, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    
    writer.writerow(["Programming Language", "Rating"])
    
    writer.writerows(data)

print(f"Scraped data saved to {filename}")
```
